```python
import numpy
import pandas as pd
import warnings 

data= 'vine_table'
warnings.filterwarnings('ignore')
```

```python
vine_df = pd.read_csv(data, index_col= False)
vine_df
```

```python
#filter total votes >= 20
filtered_df = vine_df[vine_df["total_votes"]>=20]
filtered_df
```

```python
#Filter the new DataFrame or table created in Step 1 and create a new DataFrame or 
#table to retrieve all the rows where the number of helpful_votes divided by 
#total_votes is equal to or greater than 50%.

filtered_df["votes_ratio"] = filtered_df.helpful_votes/ filtered_df.total_votes >= .5
filtered_df
```

```python
vine_verified = filtered_df[filtered_df['verified_purchase']== 'Y']
vine_verified
```

```python
vine_unverified = filtered_df[filtered_df['verified_purchase'] == "N"]
vine_unverified
```

```python
# Determine the total number of reviews, the number of 5-star reviews,
# and the percentage of 5-star reviews for the two types of review

verified_total_reviews = len(vine_verified)
unverified_total_reviews = len(vine_unverified)

verified_five_stars = (vine_verified['star_rating'].values ==5).sum()
unverified_five_stars = (vine_unverified['star_rating'].values ==5).sum()

verified_five_stars_percent = (verified_five_stars/verified_total_reviews)
unverified_five_stars_percent = (unverified_five_stars/unverified_total_reviews)
```

```python
print(verified_total_reviews, verified_five_stars, verified_five_stars_percent)
print(unverified_total_reviews, unverified_five_stars, unverified_five_stars_percent)
```
